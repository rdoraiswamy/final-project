```python
from sklearn.datasets import load_iris;
iris_dataset = load_iris();
print("Keys of iris_dataset:\n", iris_dataset.keys())
```

```python
print("Shape of Data:", (iris_dataset['data']).shape);
print("Shape of target_names:", (iris_dataset['target_names']).shape)
print("Shape of target:", (iris_dataset['target']).shape)
```

```python
print("Head of Data:", iris_dataset['data'][:5]);
```

```python
########################################################################
# Now we start the splitting
########################################################################
from sklearn.model_selection import train_test_split;
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state = 0);
```

```python
print("shape of X_train", X_train.shape);
print("shape of X_test", X_test.shape);

print("shape of y_train", y_train.shape);
print("shape of y_test", y_test.shape);
```

```python
iris_dataframe = pd.DataFrame(X_train, columns = iris_dataset.feature_names);
iris_dataframe.head
```

```python
pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(20,20)
                          , marker = 'X', hist_kwds = {'bins':20}, s=60,
                          alpha = 0.8);
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1);
knn.fit(X_train, y_train)
```

```python
type(knn)
```

```python
X_new = np.array([[5,2.9,1,0.2]]);
print("X_new shape:\n", X_new.shape)
```

```python
prediction = knn.predict(X_new);
print("prediction:", prediction);
print("Predicted target_name:", iris_dataset['target_names'][prediction])
```

```python
y_pred = knn.predict(X_test);
print("Test set Predictions for y_pred:\n", y_pred)
```

```python
print("Test set score: {:.2f}".format(np.mean( y_pred == y_test)))
```

```python
print("Test set score: {:.2f}".format(knn.score(X_test, y_test)))
```
